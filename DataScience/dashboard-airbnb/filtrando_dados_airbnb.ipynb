```
import geopandas as gpd
import pandas as pd
```

```

gdf = gpd.read_file("neighbourhoods.geojson")

gdf_proj = gdf.to_crs(epsg=31983)

gdf_proj['centroid'] = gdf_proj.geometry.centroid

gdf['centroid'] = gdf_proj['centroid'].to_crs(epsg=4326)

gdf['lon'] = gdf['centroid'].x
gdf['lat'] = gdf['centroid'].y

gdf[['neighbourhood', 'lat', 'lon']].to_csv("neighbourhood_centroids.csv", index=False, encoding='utf-8-sig')

```

```
# Abre com encoding ISO-8859-1 (ou latin1)
df_listings = pd.read_csv("listings.csv", encoding="latin1")
df_calendar = pd.read_csv("calendar.csv", encoding="latin1")

# Salva com encoding UTF-8
df_listings.to_csv("listings_utf8.csv", index=False, encoding="utf-8")
df_calendar.to_csv("calendar_utf8.csv", index=False, encoding="utf-8")
```
